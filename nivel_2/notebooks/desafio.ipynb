```python
import pandas as pd
import matplotlib.pyplot as plt

# Dicionário de faturamento
dict_faturamento = {
    'data_ref': [
        '2023-01-01', 
        '2020-02-01', 
        '2021-03-01', 
        '2022-04-01', 
        '2023-05-01',
        '2023-06-01', 
        '2020-07-01', 
        '2021-08-01', 
        '2022-09-01', 
        '2023-10-01',
        '2022-11-01', 
        '2023-12-01',
        ],
    'valor': [
        400000, 
        890000, 
        760000, 
        430000, 
        920000,
        340000, 
        800000, 
        500000, 
        200000, 
        900000,
        570000, 
        995000,
        ]
}

df = pd.DataFrame(dict_faturamento)
```

```python
media_vendas = df['valor'].mean()
print(f'Média de vendas: {media_vendas}')
```

```python
df['data_ref'] = pd.to_datetime(df['data_ref'])
df['meses_ref'] = df['data_ref'].dt.strftime('%b')
```

```python
plt.bar(x=df['meses_ref'], height=df['valor'])
```

```python
plt.plot(df['meses_ref'], df['valor'])
```
